```python
from seleniumbase import Driver
from bs4 import BeautifulSoup
import time
import csv
import logging
from selenium.webdriver.common.by import By
dic = {'A':20,'B':9,'C':13,'D':9,'E':8,'F':5,'G':9,'H':7,'I':10,'J':4,'K':7,'L':5,'M':14,'N':7,'O':4,'P':11,'Q':2,'R':9,
       'S':25,'T':12,'U':3,'V':7,'W':4,'X':1,'Y':1,'Z':2,'#':1}
# Setup logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(message)s')

# Initialize the SeleniumBase driver with options similar to headless Chrome
agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"

# Function to retrieve data for a given letter and start index
def scrape_data(letter, start_index, driver):
    url = f"https://in.indeed.com/companies/browse-companies/{letter}?start={start_index}"
    try:
        driver.get(url)
        content = driver.page_source
        if "Verifying you are human" in content:
            logging.info("CAPTCHA encountered")
            return True, None
        soup = BeautifulSoup(content, "html.parser")
        h3_texts = [tag.text for tag in soup.find_all('h3', class_='css-ry7uss e1wnkr790')]
        return False, h3_texts
    except Exception as e:
        logging.warning(f"Failed to retrieve page {url}: {e}")
        return False, None

# Configuring the web driver to work in headless mode
driver = Driver(uc=True, log_cdp=True, headless=True, no_sandbox=True, agent=agent, proxy=False)

try:
    skip_to_next = False  # Flag to skip to the next iteration
    for i in dic.keys():  # ASCII values for A-Z
        #if skip_to_next:
            #skip_to_next = False
            #continue
        
        ch_time = time.time()
        letter = i
        logging.info(f'---------------------------')
        logging.info(f'{letter=}')

        for j in range(dic[i]):
            it_time = time.time()
            start = str(j * 1000).zfill(4)

            # Scrape data for the current letter and start index
            captcha_encountered, h3_texts = scrape_data(letter, start, driver)

            if captcha_encountered:
                # Reset driver and skip to next iteration
                driver.quit()
                driver = Driver(uc=True, log_cdp=True, headless=True, no_sandbox=True, agent=agent, proxy=False)
                captcha_encountered, h3_texts = scrape_data(letter, start, driver)
                #skip_to_next = True
                #break

            if h3_texts is not None:
                # Write to CSV
                csv_filename = f'h3_texts_{letter}.csv'
                with open(csv_filename, 'a', newline='', encoding='utf-8') as file:
                    writer = csv.writer(file)
                    for text in h3_texts:
                        writer.writerow([text])

                logging.info(f'{letter}_{j} : {time.time() - it_time}')


        logging.info(f'{letter} : {time.time() - ch_time}')
        logging.info(f'---------------------------')

finally:
    # Close the browser
    driver.quit()
```
